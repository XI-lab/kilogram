# Analysis: raw counts

First we try to see what happens if we perform stupid linking by means of exact string matching, using canonical and "redirect" labels of wikipedia pages. All such labels uniquely link to a single entity (ensured by wikipedia, we don't check it here).

We aggregate the counts of the labels and also of their lowercase versions for further analysis.
Initially the script will also generate DBPedia entity dictionary which maps labels to entity types (used later):

    ./run_job.py -cmd "python dbpedia_dbm.py" -file "dbpedia_types.dbm" -m ./filter/mapper_lowercase.py -r ./filter/reducer_lowercase.py $NGRAM_PREFILTER_DIR $DBPEDIA_COUNT_DIR

Retrieve label counts:

    hdfs dfs -cat $DBPEDIA_COUNT_DIR/* > dbpedia_counts_inferred

```python
import shelve
dbpedia_types = shelve.open('../mapreduce/dbpedia_types.dbm', flag='r')
count_dict = {}
for line in open('dbpedia_counts_inferred'):
    label, values = line.split('\t')
    upper_count, lower_count = values.split(',')
    count_dict[label] = {'new_upper': int(upper_count), 'new_lower': int(lower_count),
                       'label': label, 'orig_upper': 0, 'orig_lower': 0, 'type': dbpedia_types[label][0]}
for line in open('dbpedia_counts_original'):
    label, values = line.split('\t')
    if label in count_dict:
        upper_count, lower_count = values.split(',')
        count_dict[label].update({'orig_upper': int(upper_count), 'orig_lower': int(lower_count)})
counts_df = pd.DataFrame(count_dict.values())
counts_df.head()
del count_dict
```

```python
print 'Total number of inferred types:',  counts_df[(counts_df.new_upper + counts_df.new_lower > 0)]['type'].nunique()
print 'Total number of original types:',  counts_df[(counts_df.orig_upper + counts_df.orig_lower > 0)]['type'].nunique()
```

## Inferred vs Organic capitalized/lowercase ratio plots

```python
def ratio(row):
    new_ratio = row['new_upper']/float(row['new_lower'] or 1)
    orig_ratio = row['orig_upper']/float(row['orig_lower'] or 1)
    return pd.Series({'new_ratio': new_ratio, 'orig_ratio': orig_ratio, 'type': row['type']})

ratio_df = counts_df.apply(ratio, axis=1)
print ratio_df.head()
```

```python
ax=ratio_df[(ratio_df.type=='Settlement')].plot(title='Settlement type',kind='scatter',x='orig_ratio', y='new_ratio', figsize=(18,10), xlim=(0,500), ylim=(0,500))
ax=ratio_df[(ratio_df.type=='Song')].plot(title='Song type',kind='scatter',x='orig_ratio', y='new_ratio', figsize=(18,10), xlim=(0,50), ylim=(0,50))
```

```python
ax=ratio_df.plot(kind='scatter',x='orig_ratio', y='new_ratio', figsize=(18,10), xlim=(0,1000), ylim=(0,1000))
x = np.arange(0, 1000, 0.1);
plt.plot(x, x, axes=ax, scaley=False, scalex=False)
```

```python
# Write excludes file, create lowercase mathings file
dbpediadb_lower = {}
excludes = open('dbpedia_uri_excludes.txt', 'w')
for row in counts_df.iterrows():
    row = row[1]
    exclude = False
    label = row['label']
    
    # skip uppercase
    if label.isupper():
        continue
    # skip titlecased labels than are bigrams and higher
    if label.count('_') > 0 and label.istitle():
        # add to lower db
        if row['orig_lower'] > 0:
            dbpediadb_lower[label.lower()] = label
        continue
    if row['orig_upper'] == 0:
        if row['new_upper'] == 0:
            pass
        else:
            new_ratio = row['new_upper']/float(row['new_lower'] or 1)
            if new_ratio < 20:
                exclude = True
    else:
        new_ratio = row['new_upper']/float(row['new_lower'] or 1)
        orig_ratio = row['orig_upper']/float(row['orig_lower'] or 1)
        if new_ratio < orig_ratio:
            exclude = True
    if exclude:
        excludes.write(label+'\n')
    elif row['orig_lower'] > 0:
        dbpediadb_lower[label.lower()] = label
excludes.close()        
```

```python
out = open('dbpediadb_lower.txt', 'w')
for lower_label, label in dbpediadb_lower.items():
    out.write('%s\t%s\n' % (lower_label, label))
out.close()
```

```python
from collections import defaultdict
type_counts = defaultdict(lambda: [0,0])
for i, vocab in enumerate((open('dbpedia_org_types_1gms'), open('dbpedia_infer_types_1gms'))):
    for line in vocab:
        word, count = line.split('\t')
        type_counts[word][i] = int(count)
df = pd.DataFrame(sorted(type_counts.values()), columns=['orig_types', 'inferred_types'])
```
